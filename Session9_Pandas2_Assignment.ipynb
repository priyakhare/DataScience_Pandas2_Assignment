```python

#***********************************************************************************************************

# 3.              Program : To Show The Top most preferred names in the Data from File

#***********************************************************************************************************
# Importing packages numpy and pandas and setting alias 
import numpy as np 
import pandas as pd


# Read Data from given csv File and 
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")

# Checking if any of the columns is Unnamed in all and then deleting that column 
df = df.drop(df.columns[df.columns.str.contains('unnamed',case=False)],axis=1)
#print(df.head())


# Grouping by Name , then finding the sum of the counts to get total count of Names
# Thereafter sorting the Data in descending order to get the top most preferred Names 
df_byName = df.groupby('Name')['Count'].sum().sort_values(ascending=False).head()


print('-'*80,"\nThe Top 5 Most Preferred Names are \n",'-'*80)
print(df_byName)
print('-'*80)



#***********************************************************************************************************

```

```python

#***********************************************************************************************************

#             Program : To Show The Median name occurence in the Data from File

#***********************************************************************************************************
# Importing packages numpy and pandas and setting alias 
import numpy as np 
import pandas as pd


# Read Data from given csv File and 
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")

# Checking if any of the columns is Unnamed in all and then deleting that column 
df = df.drop(df.columns[df.columns.str.contains('unnamed',case=False)],axis=1)

# To find the name , Sorting the Names to get a Sorted List of names and then getting the median 
# Getting the Id field that is unique for this median and using it to display the median name 
medianId = df.sort_values('Name',ascending=True).median().Id


print(medianId)
print("\n The Median Name occurence is : \n",'-'*80)
print(df[df['Id'] == medianId])
print('-'*80)
#***********************************************************************************************************

```

```python
#***********************************************************************************************************

#             Program : To Show The Distribution of Male, Female born count, by States for given Data 

#***********************************************************************************************************
# Importing packages numpy and pandas and setting alias 
import numpy as np 
import pandas as pd


# Read Data from given csv File and 
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")

# Checking if any of the columns is Unnamed in all and then deleting that column 
df = df.drop(df.columns[df.columns.str.contains('unnamed',case=False)],axis=1)


# Getting the Count of Female borns StateWise
df1 = pd.DataFrame(df[df['Gender'] == 'F'].groupby('State').Count.sum())

# Getting the Count of Male borns StateWise
df2 = pd.DataFrame(df[df['Gender'] == 'M'].groupby('State').Count.sum())


# Since Both DataFrames, df1, df2, have Count Column , renaming the Count accordingly
df1 = df1.rename(columns={'Count':'FemaleCnt'})
df2 = df2.rename(columns={'Count':'MaleCnt'})

# Joining both the DataFrames df1, df2 to include all the REcords , states in both DataFrames
df_Total = df1.join(df2,how="outer")

# Creating a New Column which Counts the Total Females and Males Too StateWise 
df_Total['TotalCnt'] = df_Total['FemaleCnt'] + df_Total['MaleCnt']

print('-'*80,"\nThe StateWise Distribution Of Female and Male Borns is \n",'-'*80)
print(df_Total)
print('-'*80)

#***********************************************************************************************************

```
